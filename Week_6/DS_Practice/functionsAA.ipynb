Given a number N. Print the digits of N separated by a space.

Note: Solve this problem using recursion.

Input
First line contains a number T (1 ≤ T ≤ 10) number of test cases.

Next T lines will contain a number N (0 ≤ N ≤ 109).

Output
For each test case print a single line contains the digits of the number separated by space.

Example

Input

3

121

39

123456

Output

1 2 1

3 9

1 2 3 4 5 6



```python
def print_digits(n):
    if n<10:
        print(n,end=" ")
    else:
        print_digits(n//10)
        print(n%10,end=" ")
```

```python
T=int(input())
for _ in range(T):
    n=int(input())
    print_digits(n)
    print()
```

Given a number N. Print the binary equivalent of N.

Note: Solve this problem using recursion.

Input

First line contains a number T (1 ≤ T ≤ 104) number of test cases.

Next T lines will contain a number N (1 ≤ N ≤ 109).

Output

For each test case print a single line contains the answer according to the required above.

Example

Input

2

10

3

Output

1010

11

```python
def to_binary(n):
    if n==0:
        return
    to_binary(n//2)
    print(n%2,end="")
```

```python
T=int(input())
for _ in range(T):
    n=int(input())
    to_binary(n)
    print()
```

Balanced Parentheses (Easy)

Example:

Input: (())
Output: Balanced


Input: (()(
    
Output: Not Balanced

```python
s=input()
stack=[]
balanced=True
for c in s:
    if c=='(':
        stack.append(c)

    elif c==')':
        if stack:
            stack.pop()

        else:
            balanced =False
            break


if stack:
    balanced=False


if balanced:
    print("balanced")

else:
    print("Not Balanced")



```

Write a Python program to count the occurrences of each word in a given sentence (Sample String:amr and ahmed are frindes but amr is the tallest Expected Result:2)

```python
s="amr and ahmed are frindes but amr is the tallest and and"
l=s.split()
word_count={}
for i in l:
    if i in word_count:
        word_count[i]+=1

    else:
        word_count[i]=1


print(word_count)
```

write python program to Remove keys with Values Greater than K ( Including mixed values )
nput : test_dict = {‘Gfg’ : 3, ‘is’ : 7, ‘best’ : 10, ‘for’ : 6, ‘geeks’ : ‘CS’},
K = 7
Output : {‘Gfg’ : 3, ‘for’ : 6, ‘geeks’ : ‘CS’}

```python
d = {'Gfg' : 3, 'is' : 7, 'best': 10, 'for' : 6, 'geeks': 'CS'}
f={}
for key,value in d.items():
    if isinstance(value,(int,float)):
        if value <7:
            f[key]=value

    else:
        f[key]=value


f
```
