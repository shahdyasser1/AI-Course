# 1. Opening and Closing Files

## We use the built-in open() function.

```python
# Syntax: open(filename, mode)
# Modes: r (read), w (write), a (append), x (create), b (binary), t (text)

file = open("example.txt", "w")   # open file for writing
file.write("Hello, this is my first file!\n")
file.write("Hello, this is my first file!\n")
file.write("Hello, this is my first file!\n")
file.write("Hello, from ai course\n")
file.close()  # always close files after using
```

# 2. Reading from Files

```python
file = open("example.txt", "r")
content = file.read()   # reads whole file
print(content)
file.close()
```

```python
# Read specific number of characters
file = open("example.txt", "r")
print(file.read(15))   # read first 15 characters
file.close()
```

```python
with open("example.txt", "r") as file:
    print(file.readline())
```

# 3. Appending to Files

```python
file = open("example.txt", "a")
file.write("This line is added without deleting old content.\n")
file.close()

```

# 4. Using with open() (Best Practice ✅)

## Automatically closes file when done.

```python
with open("example.txt", "r") as file:
    print(file.read())
```

```python
with open("example.txt", "r") as file:
    lines=file.readlines()
    print(lines[3])
```

# 6. Working with File Paths

# 7. Deleting Files

```python
import os

if os.path.exists("example.txt"):
    os.remove("example.txt")
    print("File deleted!")
else:
    print("File does not exist")
```

```python
with open("numbers.txt","w") as file:
  for i in range(1,21):
    file.write(f"{i}\n")
```

```python
with open("numbers.txt","r") as file:

  for i in range(1,21):
    num=int(file.readline().strip())
    if num%2==0:
      print(num)
```

```python
with open("numbers.txt","a") as file:
  file.write("End of file")
```

```python
with open("numbers.txt","r") as file1:
    content = file1.read()

with open("copy_numbers.txt","w") as file2:
    file2.write(content)
```
