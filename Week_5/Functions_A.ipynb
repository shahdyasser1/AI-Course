Write a function called count_vowels that takes a string and returns the number of vowels.

```python
def count_vowels(text):
    text=text.lower()
    vowels="aeiou"
    count=0
    for c in text:
        if c in vowels:
            count+=1
    return count
```

```python
#hello auc ahmed
count_vowels("hello auc ahmed")
```

A prime number is a number that is greater than 1 and has only two factors which are 1 and itself. The first few prime numbers are 2, 3, 5, 7, 11, 13, 17, 19, 23 and 29.

Example

Input

2

8

Don't use an array.




```python
def is_prime(n):
    if n<2:
        return False
    else:
        for i in range(2,n):
            if n%i==0:
                return False


        return True
```

Given an array A of size N. Print the minimum and the maximum number in the array.

Note: Solve this problem using function.

Input

Second line will contain N numbers (0 ≤ Xi ≤ 105).

Output
Print the minimum and the maximum number separated by a space.

Example

Input

10 13 95 1 3

```python
def max_min(l):
    mx=l[0]
    mn=l[0]
    for i in l:
        if i <mn:
            mn=i

        if i >mx:
            mx=i

    return mx,mn
```

Given an array A

Note: Solve this problem using function.

Input
First line will contain a number N
 (1≤N≤103)
 number of elements.

Second line will contain N
 numbers (0≤Ai≤103)
.

Output
Print the array after shifting right all its zeros.

Examples

Input

4

2 0 0 5

Output

2 5 0 0

Input

5

1 5 0 7 4

Output

1 5 7 4 0

```python
def shift_zeris_right(l):
    res=[]
    zeros_count=0
    for num in l:
        if num!=0:
            res.append(num)

        else:
            zeros_count+=1

    res.extend([0]*zeros_count)
    return res
```

```python
def print_digits(n):

    if n<10:
        print(n,end=" ")
        return



    print_digits(n//10)
    print(n%10,end=" ")
```


Write a function called:

def is_armstrong(n):
Requirements:

Calculate the number of digits in the given number n.

Compute the sum of each digit raised to the power of the total number of digits.

Return:

True if the number is an Armstrong number.

False otherwise.

```python
def is_armstrong(n):                #153  1^3 +5^3+3^3 =153
    num_digits=len(str(n))
    total=0
    temp=n
    while temp>0:
        digit=temp%10
        total+=digit**num_digits
        temp//=10
    return total==n
```

```python
is_armstrong(153)
```

Write a function called:

def calculator():
The function should:

Ask the user to enter two numbers.

Ask the user to choose an operation:

1 - Addition
2 - Subtraction
3 - Multiplication
4 - Division
5 - Power

Use match-case to perform the selected operation.

Print the result.

Handle division by zero properly.

If the user enters an invalid option, print "Invalid choice".

Example Run:
Enter first number: 5
Enter second number: 3
Choose operation: 1

Result = 8

```python
def calculator():
    a=float(input("Enter First Number :"))
    b=float(input("Enter Second Number :"))
    #a,b=map(int,input().split())
    print("Choose Operation :")
    print("1 - Addition")
    print("2 - Subtraction")
    print("3 - Multiplication")
    print("4 - Division")
    print("5 - Power")

    choice=int(input())

    match choice:
        case 1:
            print(a+b)

        case 2:
            print(a-b)

        case 3:
            print(a*b)

        case 4:
            if b!=0:
                print(a/b)
            else :
                print("Divide By Zero")

        case 5:
            print(a**b)


        case _:
            print("Invalid input")


```

```python
calculator()
```
