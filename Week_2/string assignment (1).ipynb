Write a program that:
Asks the user to enter a string
Prints the first character of the string
Prints the last character of the string

```python
s=input("Enter a string: ")
print(f"The first character: {s[0]}")
print(f"The last character: {s[-1]}")
```

String Methods
Write a program that:
Asks the user to enter a text
Prints the text in uppercase
Prints the text in lowercase
Prints the number of characters in the text

```python
text=input("Enter a text: ")
print(text.upper())
print(text.lower())
print(len(text))
```
