## Section 1: The Match-Case Statement (Switch)
*Introduced in Python 3.10, the `match` statement is similar to the `switch` statement in other languages. It takes an expression and compares its value to successive patterns given as one or more `case` blocks.*



### 1.1 Basic Match-Case
*Matching integer values to determine a result.*

```python
choice = 1

match choice:
    case 1:
        result = "You chose One"
    case 2:
        result = "You chose Two"
    case 3:
        result = "You chose Three"
    case _:
        result = "Invalid choice"

print(result)
```

```python
x="ai"
if x=="ai":
  print("cousre ai")
elif x=="data analysis":
  print("data")
else:
  print("non")
```

```python
x="ai"
match x:
  case "ai":
    print("course ai")
  case "data analysis":
    print("data")
  case _:
    print("non")
```

### 1.2 User Input Match
*Using `match` to convert numeric input into days of the week.*

```python
day = int(input("Enter day number (1-7): "))

match day:
    case 1: result = "Monday"
    case 2: result = "Tuesday"
    case 3: result = "Wednesday"
    case 4: result = "Thursday"
    case 5: result = "Friday"
    case 6: result = "Saturday"
    case 7: result = "Sunday"
    case _: result = "Invalid day"

print(result)
```

### 1.3 String Matching (Calculator)
*Matching string operators to perform arithmetic.*

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ") # + - * /

match operation:
    case "+": result = a + b
    case "-": result = a - b
    case "*": result = a * b
    case "/": result = a / b
    case _: result = "Invalid operation"

print(f"Result: {result}")
```

### 1.4 Guarded Cases
*Using `if` conditions inside a case check to handle ranges or specific logic.*

```python
number = int(input("Enter a number: "))

match number:
    case n if n > 0:
        result = "Positive"
    case n if n < 0:
        result = "Negative"
    case 0:
        result = "Zero"
    case _:
        result = "Unknown"

print(result)
```

```python
ls=["anas",120]
match ls:
  case [name,num] if name=="anas":
    print("name:",name)
  case [name,num]:
    print("num:",num)
```
