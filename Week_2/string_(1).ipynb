## Strings are immutable

```python
s = "Indore"
#     012345   -1 = e

print(s[0])         # I
print(s[-1])        # e


# This will NOT work:
# s[0] = 'Q'        # TypeError: strings are immutable
```

```python
print(s[::-1])      # erodnI  ← reverse
print(s[0:3])       # Ind
```

```python
# Ways to create a modified version
s = "Indore"
s = 'Q' + s[1:]
print(s)           # Qndore

# Another common pattern
s = "Indore"
s = s[:3] + 'H' + s[4:]    # Ind H re
print(s)           # IndHre
```

```python
s = 'kerolosmoheb'
print(s[::-3])     # every third character backwards
```

```python
s='abcdefgh'
s[3:6:10]
```

## More slicing examples

```python
s = "Communication"
#    0123456789ABC  --> A mean 10 , B mean 11 , c mean 12

print(s[1:4])        # omm
print(s[4:1:-1])     # umm
print(s[5:-4])       # nica
print(s[5:9]) # the same of above
print(s[::-1])       # noitacinummoC   --> reverse
print(s[1:10:2])     # ouiaiu
```

```python
print(s[4:1:1])
```

```python
name = input("Enter your name: ")
print("You entered:", name)
print("Reversed:", name[::-1])
```

## Common string methods

```python
s = "India"
print(len(s))        # 5
print(max(s))        # n
print(min(s))        # I

print("ha" * 4)      # hahahaha
```

```python
s="shahd yasser"
print(s.index("h"))
```

```python
print(s.find("h"))
```

```python
first_occurrnce=s.find("h")

print(s.find("h" , first_occurrnce + 1))
```

```python
word="banana"
target="a"
count=0
for index in range(len(word)):
    if word[index] == target:
      count+=1
      print(f"The Character {count} in index {index}")
```

```python
# f-strings (recommended way)
score = 50
print(f"Your score is {score} out of 100")

name = "youssef"
age = 23
print(f"My name is {name.upper()} and I am {age} years old.")
```

```python
# join
words = ["Hello", "from", "the", "AI", "course"]  # list
print(" ".join(words))
print("-".join(words))
print(" , ".join(words))
```

```python
# replace
text = "ali mohamed ali hussein ali"
print(text.replace("ali", "omar"))
print(text.replace("ali", "omar", 2))   # only first 2 occurrences
```

```python
# case conversion
city = "cairo"
print(city.upper())
print(city.capitalize())

shout = "WELCOME TO PYTHON"
print(shout.lower())
```

```python
# split
sentence = "python is awesome and fun"
print(sentence.split())           # default = split on spaces

data = "name,age,city"
print(data.split(","))

msg = "hello***world***python"
print(msg.split("***"))
```
