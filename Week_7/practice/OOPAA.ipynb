Question 1: Encapsulation

Create a class BankAccount with the following requirements:

A private attribute __balance that stores the account balance.

A public method deposit(amount) to add money to the balance.

A public method withdraw(amount) to remove money if there is enough balance.

A public method get_balance() that returns the current balance.

Input/Output Example:

acc = BankAccount()

acc.deposit(500)

acc.withdraw(200)

print(acc.get_balance())  # Output: 300

```python
class BankAccount:
    def __init__(self):
        self.__balance=0

    def deposite(self,amount):
        if amount >0:
            self.__balance+=amount

    def withdraw(self,amount):
        if 0<amount<=self.__balance:
            self.__balance-=amount


    def get_balance(self):
        return self.__balance


```

```python
class Student:
    def __init__(self,grade):
        self.grade=grade

    def set_grade(self,new_grade):
        self.grade=new_grade

    def get_grade(self):
        return self.grade

    def __str__(self):
        return f"{self.grade}"

    def __add__(self, other):
        return self.grade+other.grade

```

## Hospital Management System

```python
from abc import ABC ,abstractmethod
class Person(ABC):
    def __init__(self,name,age):
        self.name=name
        self.age=age


    @abstractmethod
    def print_detials(self):
        pass




```

```python
class Doctor(Person):
    def __init__(self,name,age,specialty):
        super().__init__(name,age)
        self.specialty=specialty
        self.Patients=[]

    def print_detials(self):
        print(f"doctor name is {self.name} doctor age is {self.age} specialty is {self.specialty}")
```

```python
class Patient(Person):
    def __init__(self,name,age,symptoms):
            super().__init__(name,age)
            self.symptoms=symptoms
            self.assigned_doctor=None


    def print_detials(self):
          print(f"patient name is {self.name} and age is {self.age} symptoms is {self.symptoms}")
```

```python
class Hospital:
    def __init__(self):
        self.hospital_doctors=[]
        self.hospital_Patients=[]


    def add_doctor(self,doctor):
        self.hospital_doctors.append(doctor)


    def add_patient(self,patient):
        self.hospital_Patients.append(patient)
        self.assign_doctor(patient)



    def assign_doctor(self,patient):
        if not self.hospital_doctors:
            print("No Doctors")
            return

        temp_doctor=self.hospital_doctors[0]
        for d in self.hospital_doctors:
            if len(d.Patients)<len(temp_doctor.Patients):
                temp_doctor=d

        temp_doctor.Patients.append(patient)
        patient.assigned_doctor=temp_doctor

        print(f"patient {patient.name} assigned to doctor {temp_doctor.name}")




```
