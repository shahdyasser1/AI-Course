## Section 1: Encapsulation
*Encapsulation is the bundling of data (attributes) and methods (functions) into a single unit (class). It also involves restricting direct access to some of an object's components (using private attributes) to protect data integrity.*



### 1.1 Private Attributes & Getters/Setters
*Using double underscores `__` to make attributes private, and methods to access them securely.*

```python
# Example 1: Bank Account System
class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number  # Private attribute
        self.__balance = balance                # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}")
        else:
            print("Deposit amount must be positive!")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}")
        else:
            print("Insufficient balance or invalid amount!")

    def get_balance(self):
        return self.__balance  # Getter method to safely access private data
```

```python
# Using the BankAccount class
my_account = BankAccount("12345678", 5000)
my_account.deposit(2000)
my_account.withdraw(1000)

# Accessing balance via getter
print(f"Current Balance: ${my_account.get_balance()}")

# Trying to access private attribute directly will fail
# print(my_account.__balance)  # This would raise an AttributeError
```

### 1.2 Encapsulation with Logic
*Encapsulation allows us to validate data before changing it.*

```python
# Example 2: Employee Salary Management
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary  # Private salary

    def set_salary(self, new_salary):
        if new_salary > self.__salary:
            self.__salary = new_salary
            print(f"Salary updated for {self.name}.")
        else:
            print("New salary cannot be lower than current salary.")

    def get_salary(self):
        return self.__salary

emp = Employee("Youssef", 5000)
emp.set_salary(6000)  # Valid update
print(f"Current Salary: {emp.get_salary()}")
```

## Section 2: Abstraction
*Abstraction hides complex implementation details and exposes only the essential features. In Python, this is achieved using Abstract Base Classes (ABC). An abstract class cannot be instantiated directly; it serves as a blueprint for other classes.*



### 2.1 Abstract Base Classes (Shapes)
*Defining a template where children MUST implement specific methods.*

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass
```

```python
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius
```

```python
# Using Abstract Implementations
rect = Rectangle(5, 10)
circle = Circle(7)

print(f"Rectangle Area: {rect.area()}")
print(f"Circle Area: {circle.area()}")
```

### 2.2 Abstraction in Real World (Animals)
*Enforcing consistent behavior (like making sound or eating) across different species.*

```python
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

class Dog(Animal):
    def make_sound(self):
        return "Woof!"

animals = [Cat(), Dog()]
for animal in animals:
    print(animal.make_sound())
```

## Section 3: Inheritance
*Inheritance allows a class (child) to acquire the properties and methods of another class (parent). It promotes code reusability.*



### 3.1 Single Inheritance
*One child inherits from one parent.*

```python
class BaseAnimal:
    def __init__(self, name):
        self.name = name

    def sleep(self):
        return f"{self.name} is sleeping."

class Lion(BaseAnimal):
    def roar(self):
        return "ROAR!"

simba = Lion("Simba")
print(simba.sleep())
print(simba.roar())
```

```python
class A:

    def add(self):
        print("Add from class A")

class B(A):

    def sub(self):
        print("Sub from class B")

class C(B):

    def show(self):
        print("Show from class C")

z = C()
z.show()
z.sub()
z.add()
```

### 3.2 Multilevel Inheritance
*Inheritance chaining: Child -> Parent -> Grandparent.*

```python
A
|
B
|
C
```

```python
# Grandparent
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

# Parent
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

# Child
class ElectricCar(Car):
    def __init__(self, brand, model, battery):
        super().__init__(brand, model)
        self.battery = battery

    def details(self):
        print(f"{self.brand} {self.model} with {self.battery}kWh battery")

tesla = ElectricCar("Tesla", "Model S", 100)
tesla.details()
```

### 3.3 Multiple Inheritance
*A child inherits from multiple parents simultaneously.*

```python
class Engine:
  def __init__(self):
    print("engine")
  def start_engine(self):
        print("Engine started.")


class Radio:
    def __init__(self):
      print("radio")
    def play_music(self):
        print("Playing music.")

class SmartCar(Engine, Radio):
    def __init__(self):
        super().__init__()
        Radio().__init__()
        print("smart car")
    def drive(self):
        print("Driving...")

my_car = SmartCar()
my_car.start_engine()
my_car.play_music()
my_car.drive()
```

## Section 4: Polymorphism
*Polymorphism means "many forms". It allows objects of different classes to be treated as objects of a common superclass. The most common implementation is Method Overriding.*



### 4.1 Method Overriding (Birds Example)
*Different classes providing their own implementation of the same method.*

```python
class Bird:
    def fly(self):
        return "Some birds can fly."

class Sparrow(Bird):
    def fly(self):
        return "I fly at low altitudes."

class Eagle(Bird):
    def fly(self):
        return "I soar high in the sky."

class Penguin(Bird):
    def fly(self):
        return "I cannot fly, but I swim!"

# Polymorphism in action
birds = [Sparrow(), Eagle(), Penguin()]

for bird in birds:
    print(bird.fly())
```

### 4.2 Polymorphism in Payment Systems
*Treating Credit Cards and PayPal objects identically using a common interface.*

```python
class CreditCard:
    def pay(self, amount):
        print(f"Paying ${amount} using Credit Card.")

class PayPal:
    def pay(self, amount):
        print(f"Paying ${amount} using PayPal.")

def process_payment(payment_method, amount):
    # The function doesn't care which object it is, as long as it has a .pay() method
    payment_method.pay(amount)
###########################################################
process_payment(CreditCard(), 100)
process_payment(PayPal(), 50)
```
