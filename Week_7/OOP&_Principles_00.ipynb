## Section 3: Instance Methods
*Methods are functions defined inside a class that describe the behaviors of an object. They usually take `self` as the first parameter to access the object's attributes.*

### 3.1 Defining Behaviors
*Adding actions like displaying info or calculating data.*

```python
# Example 1: Car Display Info
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        print(f"This car is a {self.brand} {self.model}")

c1 = Car("Toyota", "Corolla")
c1.display_info()
```

```python
# Example 2: Bank Account (Logic in Methods)
class BankAccount:
    def __init__(self, owner, balance=0):  # contructor with defualt vlaues
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New Balance: ${self.balance}")

```

```python
account = BankAccount("Sara", 100)
account.deposit(50)
```

```python
account_1=BankAccount("nour",200)
account_1.deposit(300)
```

```python
# Example 3: Car Engine Logic (State Change)
class SmartCar:
    def __init__(self):   # defualt contructor
        self.is_running = False

    def start_engine(self):
        if not self.is_running:
            self.is_running = True
            print("Engine started.")
        else:
            print("Engine is already running.")

my_smart_car = SmartCar()
my_smart_car.start_engine()
```

## Section 4: Class Attributes vs Instance Attributes
* **Instance Attributes:** Specific to one object (`self.name`).
* **Class Attributes:** Shared by ALL objects of that class (`Species = "Human"`).



### 4.1 Shared Data
*Using class attributes to store common properties.*

```python
# Example 1: Car Wheels (Shared Attribute)
class Car:
    wheels = 4  # Class Attribute

    def __init__(self, brand):
        self.brand = brand  # Instance Attribute

c1 = Car("BMW")
c2 = Car("Audi")

print(f"{c1.brand} has {c1.wheels} wheels")
print(f"{c2.brand} has {c2.wheels} wheels")
```

```python
print(f"{c2.brand} has {Car.wheels} wheels")
```

```python
# Example 2: Employee Company (Shared Attribute)
class Employee:
    company_name = "TechCorp"

    def __init__(self, name):
        self.name = name

```

```python
emp1 = Employee("Ahmed")
print(f"{emp1.name} works at {Employee.company_name}")
```

## Section 5: Types of Methods (Instance, Class, Static)
* **Instance Method:** Accesses `self` (can change object state).
* **Class Method:** Accesses `cls` (can change class state). Uses `@classmethod`.
* **Static Method:** Independent helper function. Uses `@staticmethod`.*



### 5.1 Advanced Method Implementation
*Combining all three types in one class.*

```python
class Car:
    wheels = 4

    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    # Instance Method: Accesses specific object data
    def display_info(self):
        print(f"{self.brand} made in {self.year}")

    # Class Method: Accesses class variables
    @classmethod
    def change_wheels(cls, new_count):
        cls.wheels = new_count
        print(f"All cars now have {cls.wheels} wheels")

    # Static Method: Logical helper, no access to self or cls
    @staticmethod
    def is_old(year):
        if year < 2020:
            return "Old Model"
        return "Modern Model"
```

```python
# Testing the Methods
my_car = Car("BMW", 2019)

my_car.display_info()                 # Instance Call
Car.change_wheels(3)                  # Class Call
print(Car.is_old(my_car.year))        # Static Call
```

## Section 6: Practice Problem
*Create a Student class that tracks individual student data and a global student count.*

### 6.1 Student Class Solution
*Attributes: Name, Age, Courses. Class Attribute: Student Counter.*

```python
class Student:
    # Class attribute to count students
    student_count = 0

    def __init__(self, name, age, courses):
        self.name = name
        self.age = age
        self.courses = courses
        # Increment class attribute whenever a new object is created
        Student.student_count += 1

    def display_student(self):
        print(f"Student: {self.name}, Age: {self.age}, Courses: {self.courses}")
        print(f"Total Students Registered: {Student.student_count}")
```

```python
# Creating Students
s1 = Student("Ahmed", 25, 6)
s2 = Student("Sara", 22, 5)

s1.display_student()
s2.display_student()
```
