## Section 1: Introduction to Functions
*Functions are reusable blocks of code that perform a specific task. They help organize code, avoid repetition, and make programs easier to test.*



### 1.1 Defining and Calling Functions
*Using the `def` keyword to create a function.*

```python
def greet():
    print("Hello user")
# Calling the function
greet()
```

### 1.2 Function Arguments
*Passing data into functions using parameters.*

```python
def greet_person(name):
    print(f"Hello {name}")

greet_person("Shahd")
greet_person("Anas Emad")
```

```python
greet_person(20)
greet_person("Anas Emad")
```

```python
def add(x, y):
    print(x + y)

add(5, 6)
add(20,30)
```

### 1.3 Return vs. Print
*`print` displays a value to the console, but `return` sends a value back to the code that called the function so it can be stored in a variable.*

```python
# Function that only prints (returns None)
def add_print(x, y):
    print(x + y)

# Function that returns a value
def add_return(x, y):
    return x + y

x = add_print(5, 6)   # Output: 11
print(x)              # Output: None
add_print(10,2)

y = add_return(5, 6)  # No Output yet
print(y)              # Output: 11
```

### 1.4 Logic Inside Functions
*Encapsulating control flow (if/else, loops) inside functions.*

```python
def check_number(num):
    if num > 0:
        print("Positive number")
    else:
        print("Negative number")

check_number(5)
check_number(-5)
```

```python
def even_odd_checker(numbers):
    for i in numbers:
        if i % 2 == 0:
            print(f"Even number {i}")
        else:
            print(f"Odd number {i}")

even_odd_checker([1, 2, 3, 4, 5, 6, 7, 8])
```

## Section 2: Advanced Arguments
*Python offers flexible ways to pass arguments, including keyword arguments, defaults, and variable-length arguments.*



### 2.1 Keyword and Default Arguments
*Keyword arguments allow you to pass values by parameter name. Default arguments provide a fallback value if none is passed.*

```python
def info(name, age):
    print(f"Hello {name} , age: {age}")

# Positional
info("Shahd", 20)
# Keyword (Order doesn't matter)
info(age=20, name="Shahd")
```

```python
# Default Parameter
def discount(price, dis=0.15):
    return price * dis

print(discount(90))       # Uses default 0.15
print(discount(90, 0.1))  # Uses provided 0.1
```

```python
def circle_info(radius):
    area = 3.14 * radius * radius
    diameter = 2 * radius
    return area, diameter
a, d = circle_info(6)  # calling
print("Arae= ",a,"  ","Diameter= ", d)
```

### 2.2 Variable Length Arguments (*args)
*Allows a function to accept any number of positional arguments. Internally, `*args` is a tuple.*

```python
def summation(*args):
    return sum(args)

print(summation(19, 5, 8, 8, 8))
print(summation(19, 5, 8, 8, 8,20,90,30,100))
```

### 2.3 Variable Key-Value Arguments (**kwargs)
*Allows a function to accept any number of keyword arguments. Internally, `**kwargs` is a dictionary.*

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"Key: {key}, Value: {value}")

print_info(name="shahd", Age=20, City="Cairo",phone="01023345")
#print_info("shahd",20,"eg")  #Error
```

```python
print_info(company="INSTANT",location="dokki",time=11)   # key , value
```

## Section 3: Recursion
*Recursion is a technique where a function calls itself to solve a problem. It requires a 'base case' to stop the loop.*



### 3.1 Factorial Example
*Calculating 5! (5 * 4 * 3 * 2 * 1) using recursion.*

```python
def factorial(num):
    if num == 0 or num == 1:
        return 1
    else:
        return num * factorial(num - 1)

print(factorial(5))
```

```python
def gen(n):
  if n==0:
    return
  else:
    gen(n-1)
    print(n)
```

```python
gen(5)
```

### 3.2 Countdown Example
*Printing numbers in descending order.*

```python
def count_down(num):
    if num == 1:
        return 1
    else:
        print(num)
        count_down(num - 1)

count_down(5)
```

## Section 4: Lambda Functions
*Lambda functions are small, anonymous functions defined with the `lambda` keyword. They are often used with functions like `map` and `filter`.*

### 4.1 Basic Syntax
*`lambda arguments: expression`*

```python
def add(x,y):
  return x+y
```

```python
add = lambda x, y : x + y
print(add(5, 6))
```

### 4.2 Using Map with Lambda
*Applying a function to every item in a list.*

# map = transformation

# "Take each value… apply a function to it… and return the transformed value."

```python
prices = [100, 1000, 5000]

# Apply 10% tax/fee to all prices
new_prices = list(map(lambda x: x * 0.1, prices))
print(new_prices)
```

```python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x * 2 , nums))
print("result:",result)
print("nums:",nums)
```

# filter = selection

# "Take each value… check a condition… and return only the values that pass the condition."

```python
nums = [1, 2, 3, 4]

result = list(filter(lambda x: x % 2 == 0, nums))
print(result)
```

# Functions as an Argument

```python
def subSum(x,y):
  return x + y

def mul(x,y):
  return x * y


def app(function ,  x, y):
  return function(x,y)



#calling
print(app(subSum, 5, 10))
print()
print(app(mul, 5, 10))
```

```python
nums = [10, 50, 30, 80, 60]

largest = nums[0]
second = nums[0]

for n in nums:

    if n > largest:
        second = largest
        largest = n

    elif n > second and n != largest:
        second = n

print(second)
```
