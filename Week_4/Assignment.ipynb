Q6.
Write a program that takes a month number (1–12) and prints(use if and match):

"Winter"

"Spring"

"Summer"

"Autumn"

````python
month=int(input("Enter a Month number(1–12): "))
if month>=6 and month<=8:
  print("Summer")
elif month>=3 and month <=5:
  print("Spring")
elif month>=9 and month <=11:
  print("Autumn")
elif month in (1,2,12):
  print("Winter")
else:
  print("Invalid month")
````

````python
match month:
  case 6|7|8:
    print("Summer")
  case 3|4|5:
    print("Spring")
  case 9|10|11:
    print("Autumn")
  case 1|2|12:
    print("Winter")
  case _:
    print("Invalid month")
````

Q7.
Take a character from the user(use if and match).
Check if it is a vowel (a, e, i, o, u) or not.

````python
c=input("Enter a character: ").lower()
vowels="aeiou"
if c in vowels:
  print("vowel")
else:
  print("not vowel")
````

````python
match c:
  case 'a'|'o'|'e'|'i'|'u':
    print("vowel")
  case _:
    print("not vowel")
````

Write a program that:

Checks if a number is prime using a for loop

````python
num=int(input("Enter a number: "))
is_prime=1
for i in range(2,num):
  if num % i==0:
    is_prime=0
    break

if is_prime:
  print("Prime")
else:
  print("Not prime ")
````

Given a number N. Print a pyramid that has N rows.

For more clarification see the example below.

Input

Only one line containing a number N (1 ≤ N ≤ 99).

Output

Print the answer according to the required above.

Note
Don't print any extra spaces after symbol " * ".

Input

4

Output


```
   *
  ***
 *****
*******



```


````python
n = int(input())
for i in range(1,n+1):
    spaces = n-i  # 4,3,2,1,0
    stars = 2*i-1 # 1,3,5,7,9
    print(" " * spaces + "*" * stars)
````

Given a number N. Print a diamond that has 2N rows.

For more clarification see the example below.

Input
Only one line containing number N (1 ≤ N ≤ 99).

Output
Print the answer according to the required above.

Example
Input

4

Output
```
   *
  ***
 *****
*******
*******
 *****
  ***
   *
```
Note

Don't print any extra spaces after symbol " * ".

````python
n=int(input())
for i in range(1,n+1):
    spaces = n-i  # 4,3,2,1,0
    stars = 2*i-1 # 1,3,5,7,9
    print(" " * spaces + "*" * stars)

for i in range(n,0,-1):
    spaces = n-i  # 0,1,2,3,4
    stars = 2*i-1 # 9,7,5,3,1
    print(" " * spaces + "*" * stars)
````
