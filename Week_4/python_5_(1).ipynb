## Section 1: The `while` Loop
*Unlike `for` loops which iterate over a sequence, `while` loops run as long as a condition is `True`. They are useful when we don't know beforehand how many times the loop needs to run.*



### 1.1 Basic Syntax
*A simple counter using a `while` loop.*

```python
i = 0

while i <= 10:
    print(f"Current count: {i}")
    i += 1  # Increment to prevent infinite loop
```

```python
print(i)
```

### 1.2 Logic Inside Loops
*Categorizing numbers as they are generated.*

```python
i = 0
while i <= 10:
    if i % 2 == 0:
        print(f"Even number: {i}")
    else:
        print(f"Odd Number: {i}")
    i += 1
```

### 1.3 Handling Negative Ranges
*Loops can handle negative starting points and classify data.*

```python
i = -5

while i <= 5:
    if i < 0:
        print(f"Negative Number: {i}")

    else:
        print(f"Positive Number: {i}")

    i += 1
```

## Section 2: Loop Control (`break` and `continue`)
*Controlling the flow of the loop manually.*

### 2.1 The `break` Statement
*Stops the loop immediately when a condition is met.*

```python
i = 0

while i <= 10:
    if i == 5:
        print("Breaking loop at 5")
        break # Exit loop immediately
    print(i)
    i += 1
```

### 2.2 The `continue` Statement
*Skips the rest of the code in the current iteration and jumps to the next one.*

```python
i = 0

while i <= 10:
    i += 1
    if i == 5:
        continue # Skip printing 5

    print(i)
```

## Section 3: Real World Application (Authentication)
*Using `while True` to create a persistent prompt until valid data is entered.*



### 3.1 Username Verification
*Asking for input repeatedly until the correct username is given.*

```python
valid_user = "admin"

while True:
    username_input = input("Enter your username: ")
    if username_input == valid_user:
        print("Success")
        break

    else:
        print("Invalid username, try again.")
```

### 3.2 Full Login System
*Checking both username and password.*

```python
valid_user = "admin"
valid_pass = "admin123"

while True:
    u_in = input("Enter username: ")
    p_in = input("Enter password: ")

    if u_in == valid_user and p_in == valid_pass:
        print("Login Successful")
        break
    else:
        print("Invalid credentials, please try again.")
```

## Section 4: Algorithmic Thinking (Digit Manipulation)
*Using `while` loops to process numbers mathematically (extracting digits).*



### 4.1 Counting Digits
*Using integer division `// 10` to reduce the number.*

```python
num = 153
count = 0

while num > 0:
    num //= 10   # Remove the last digit
    count += 1

print(f"Number of digits: {count}")
```

### 4.2 Sum of Digits
*Using modulo `% 10` to get the last digit and add it to a total.*

```python
print(123%10)
```

```python
print(123//10)
```

```python
num = 123
total = 0

while num > 0:
    digit = num % 10   # Get last digit
    total += digit
    num //= 10         # Remove last digit

print(f"Sum of digits: {total}")
```

### 4.3 Reversing a Number
*Constructing a new number by reversing digits.*

```python
num = 456
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

print(f"Reversed number: {reverse}")
```

### 4.4 Palindrome Checker
*Checking if a number reads the same backward as forward.*

```python
num = 121
temp = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if temp == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

## Section 5: Complex Algorithms
*Combining loops and math for advanced logic problems.*

### 5.1 Decimal to Binary Conversion
*Converting a standard integer to its binary string representation.*

```python
num = 10
binary = ""

while num > 0:
    binary = str(num % 2) + binary   # Prepend the remainder
    num //= 2

print(f"Binary: {binary}")
```

### 5.2 Armstrong Number Checker
*An Armstrong number is a number that is equal to the sum of digits raised to the power of the number of digits (e.g., 153).*

```python
num = 153
temp = num
digits = 0
sum_of_powers = 0

# Step 1: Count digits
while temp > 0:
    temp //= 10
    digits += 1     # Or-->   digits = len(str(num))

# Step 2: Calculate sum of powers
temp = num
while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** digits
    temp //= 10

# Step 3: Compare
if sum_of_powers == num:
    print(f"{num} is an Armstrong number")
else:
    print(f"{num} is not an Armstrong number")
```

## Section 6: Nested Loops (Recap)
*Iterating through lists of lists.*



```python
numbers = [[1, 2],
           [3, 4],
           [5, 6]]

print("Flattening list:")
for sublist in numbers:
    for item in sublist:
        print(item)
```
