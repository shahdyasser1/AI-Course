Exercise 6 (Easy):
Write a program that takes a number from 1 to 7 and prints the corresponding day of the week using match-case:
1 → Monday, 2 → Tuesday, …, 7 → Sunday

```python
day=int(input())
match day:
    case 1:
        print("Monday")

    case 2:
        print("Tuesday")

    case 3:
        print("Wedensday")


    case 4:
        print("Thurasday")



    case 5:
        print("Friday")



    case 6:
        print("Saturday")

    case 7:
        print("Sunday")

    case _:
        print("Invalid Number")
```

Exercise 7 (Easy-Medium):
Ask the user to input a single letter and print:

"Vowel" if it is a vowel (a, e, i, o, u)

"Consonant" if it is not a vowel

```python
letter=input().lower()
match letter:
    case 'a':
        print("Vowel")
    case 'e':
        print("Vowel")
    case 'i':
        print("Vowel")
    case 'o':
        print("Vowel")
    case 'u':
        print("Vowel")
    case _:
        print("Not Vowel")
```

Exercise 9 (Medium-Hard):
Write a simple calculator:

Ask the user to input an operator (+, -, *, /) and two numbers.

Use match-case to perform the operation and print the result.

```python
a=int(input())
b=int(input())
op=input()
match op:
    case '+':
        print(a+b)
    case '-':
        print(a-b)
    case '*':
        print(a*b)
    case '/':
        if b==0:
            print("Invalid div by zero")

        else:
            print(a/b)

    case _:
        print("Invalid Operator")
```
