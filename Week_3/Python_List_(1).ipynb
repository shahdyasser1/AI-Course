**Extend**

```python
# Example 3: extend() combines another list into the current one
more_colors = ["Purple", "Orange"]
colors.extend(more_colors)
print(colors)
```

```python
n = [2,4,5,3,6]
n.extend([1,2,3])
print(n)
```

pop([index])

```python
my_list = [1, 2, 3,4,5]
a=my_list.pop()

print(a)
print(my_list)
```

#copy()

Returns a shallow copy of the list.

```python
my_list = [1, 2, 3]
new_list = my_list.copy()

print(new_list)  # Output: [1, 2, 3]
my_list.append(4)
print(new_list)   # Output: [1, 2, 3]
print(my_list)

print(id(my_list[0]))
print(id(new_list[0]))
```
